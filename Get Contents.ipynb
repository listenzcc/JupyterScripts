# Read Content Resources of National Population Survey

The Toolkit is **Beautiful Soup**

```python
import re 
import urllib
from bs4 import BeautifulSoup
```

```python
url = 'http://www.stats.gov.cn/tjsj/pcsj/rkpc/6rp/left.htm'
```

## Read Raw Content

```python
resp = urllib.request.urlopen(url)
content = resp.read()
content
```

## Setup Soup Parser

```python
soup = BeautifulSoup(content, 'html.parser')
soup
```

## Parse Content Resources

```python
c = re.compile('html/[ABf].*\.htm')
lst = soup.find_all('a', attrs={'href': c})
lst
```

```python
resources = dict()
for e in lst:
    resources[e.attrs['href']] = e.text
resources
```
