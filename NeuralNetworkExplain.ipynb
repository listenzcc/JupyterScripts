```python
import pandas as pd 
import plotly.express as px
import numpy as np
```

```python
xNum = 100
yNum = 200
```

```python
xAxis = np.linspace(-2, 2, xNum, endpoint=False)
yAxis = np.linspace(-2, 2, yNum, endpoint=False)

grid = np.meshgrid(xAxis, yAxis)
grid
```

```python
def sigmoid(x):
    return np.exp(x) / (1 + np.exp(x))

def transform(x, y, cs1, cs2):
    vs = []
    for c in cs1:
        vs.append(sigmoid((x-c[2])) * c[0] + sigmoid((y-c[3])) * c[1])
    vs = np.array(vs)
    
    return sigmoid(np.sum(vs * cs2))
```

```python
lst = []

n = 3
cs1 = np.random.randn(n, 4)
cs2 = np.random.randn(n)

for x, y in zip(grid[0].flatten(), grid[1].flatten()):
    z = transform(x, y, cs1, cs2)
    lst.append((x, y, z))
    
df = pd.DataFrame(lst, columns=['x', 'y', 'v'])

fig = px.scatter(df, x='x', y='y', color='v')
fig.show()

df
```
