```python
import time
import threading
```

```python
class Timer(object):
    def __init__(self):
        self.reset()
        
    def reset(self):
        self.t = time.time()
        
    def elapsed(self, quite=False):
        d = time.time() - self.t
        if not quite:
            print(d)
        return d
```

```python
num = 100
```

```python
interval = 40 # ms
```

```python
timer = Timer()

for j in range(num):
    print('{}-{:0.2f}-{:0.4f}'.format(j, j*interval/1000, timer.elapsed(quite=True)))
    time.sleep(interval / 1000)
```
